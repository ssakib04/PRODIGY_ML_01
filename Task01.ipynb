```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("house_price_regression_dataset.csv")
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['Square_Footage', 'Num_Bedrooms', 'Num_Bathrooms']], df['House_Price'])
```

```python
%matplotlib inline
#the relationship between 'square footage' and 'price'
plt.xlabel('Square Footage')
plt.ylabel('Price')
plt.scatter(df["Square_Footage"], df["House_Price"], marker='+', color='blue')

plt.plot(df["Square_Footage"], reg.predict(df[["Square_Footage", "Num_Bedrooms", "Num_Bathrooms"]]), color='red')
plt.show()
```

```python
input_data = pd.DataFrame([[5000, 2, 2]], columns=["Square_Footage", "Num_Bedrooms", "Num_Bathrooms"])#These are the 'Square Footage', 'Number of Bedrooms' and 'Number of Bathrooms' respctively that the model takes as input to give the predicted 'Price' as output
reg.predict(input_data)
```
